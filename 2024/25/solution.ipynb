```python
from collections import *
from heapq import *
from functools import *
from math import *
import numpy as np
import copy
import time

data = None
raw_templates= []
raw_keys = []


with open('input.txt', 'r') as file:
    data = file.readlines()
    data = [d.strip() for d in data]
    for i in range(0,len(data),8):
        schematic = data[i:i+7]
        if '#' in schematic[0]:
            raw_templates.append(schematic)
        else:
            raw_keys.append(schematic)

templates = []
keys = []

for template in raw_templates:
    tmp = [-1 for i in range(5)]
    for row in template:
        for idx,cell in enumerate(row):
            if cell == '#':
                tmp[idx] += 1
    templates.append(tuple(tmp))
for key in raw_keys:
    tmp = [-1 for i in range(5)]
    for row in key:
        for idx,cell in enumerate(row):
            if cell == '.':
                tmp[idx] += 1
    keys.append(tuple(tmp))

print(templates,keys)
print(len(templates),len(keys))
templates = set(templates)
keys = set(keys)
print(len(templates),len(keys))
common = templates & keys
print(common)
print(len(common))

```
