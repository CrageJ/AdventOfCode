```python
class Pad:
    def __init__(self,pad):
        self.pad = pad
        self.x,self.y = -1,-1
        for y,row in enumerate(pad):
            for x,cell in enumerate(row):
                if cell == 'A':
                    self.x,self.y = x,y
    def commands(self,cmds):
        f = [c for cmd in cmds for c in self.move(cmd) if c != '']
        if 'E' in f:
            return ['E']
        return f
    def move(self,c):
        if c == 'E':
            return 'E'
        if not (0 <= self.x < len(self.pad[0]) and 0 <= self.y < len(self.pad) ):
            return 'E'
        if c == "A":
            return self.pad[self.y][self.x]
        y,x = self.y,self.x
        for d,(dx,dy) in zip("^>v<",[(x,y+1),(x+1,y),(x,y-1),(x-1,y)]):
            if d == c:
                self.x,self.y = dx,dy
                if 0 <= self.x < len(self.pad[0]) and 0 <= self.y < len(self.pad) :
                    return ''
                return 'E'
        return 'E'

class NumPad(Pad):
    def __init__(self):
        super().__init__([["-1","0","A"],["1","2","3"],["4","5","6"],["7","8","9"]])
class ArrowPad(Pad):
    def __init__(self):
        super().__init__([["<","v",">"],[" ","^","A"]])

class Conga:
    def __init__(self):
        self.line=[ArrowPad(),ArrowPad(),NumPad()]
    def passMessage(self,msg):
        msg = [c for c in msg]
        for pad in self.line:
            msg = pad.commands(msg)
        return msg


c = Conga()
a = ''.join(c.passMessage("<<vAA>A>^AAvA<^A>AAvA^A<vA>^A<A>A<vA>^A<A>A<<vA>A>^AAvA<^A>A"))
print(a)
```

```python
class Pad:
    def __init__(self,pad):
        self.pad = pad
        self.x,self.y = -1,-1
        for y,row in enumerate(pad):
            for x,cell in enumerate(row):
                if cell == 'A':
                    self.x,self.y = x,y
    def commands(self,cmds):
        f = [c for cmd in cmds for c in self.move(cmd)]
        if 'E' in f:
            return ['E']
        return f
    def move(self,c):
        for y in range(len(self.pad)):
            for x in range(len(self.pad[0])):
                cell = self.pad[y][x]
                if cell == c:
                    ydiff = y-self.y
                    xdiff = x-self.x
             
                    cmds = []
                    if xdiff > 0:
                        for _ in range(xdiff):
                            cmds.append(">")
                    else:
                        for _ in range(-xdiff):
                            cmds.append("<")
                    if ydiff > 0:
                        for _ in range(ydiff):
                            cmds.append("^")
                    else:
                        for _ in range(-ydiff):
                            cmds.append("v")
                    cmds.append("A")
          
                    self.x = x
                    self.y = y
                    return cmds
class NumPad(Pad):
    def __init__(self):
        super().__init__([["-1","0","A"],["1","2","3"],["4","5","6"],["7","8","9"]])
class ArrowPad(Pad):
    def __init__(self):
        super().__init__([["<","v",">"],[" ","^","A"]])

class Conga:
    def __init__(self):
        self.line=[NumPad(),ArrowPad(),ArrowPad()]
    def passMessage(self,msg):
        msg = [c for c in msg]
        for pad in self.line:
            msg = pad.commands(msg)
        return msg

sm = 0
for msg in commands:
    c = Conga()
    mmm = c.passMessage(msg)
    msglen = len(mmm)
    numeric = [c for c in msg if c != "A"]
    numeric = int(''.join(numeric))
    complexity= numeric*msglen
    print(msg,msglen,numeric,complexity,''.join(mmm))
    sm += complexity
print(sm)
```
