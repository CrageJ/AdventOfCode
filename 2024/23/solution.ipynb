```python
from collections import *
from heapq import *
from functools import *
from math import *
import numpy as np
import copy
import time

data = None

with open('input.txt', 'r') as file:
    data = file.readlines()
    data = [d.strip().split('-') for d in data]
#print(data)
```

```python
adj = defaultdict(set)
tset = set()
computerset = set()
for a,b in data:
    #print(a,b)
    adj[a].add(b)
    adj[b].add(a)
    computerset.add(a)
    computerset.add(b)
    if a[0] == 't':
        tset.add(a)
    if b[0] == 't':
        tset.add(b)

path_nodes = {} # node to depth
path_depth = {} # depth to node
sets = set()
def dfs(node,depth):
    if node in path_nodes:
        if path_nodes[node] == depth-3:
            #print('here')
            nodes = sorted(path_depth[d] for d in range(depth-3,depth))
            sets.add(tuple(nodes))
        return
    if depth == 3:
        return
    path_nodes[node] = depth
    path_depth[depth] = node
    for nnode in adj[node]:
        dfs(nnode,depth+1)
    del path_nodes[node]
    del path_depth[depth]
    return

for c in tset:
    dfs(c,0)
#print(sets)
tset = [(a,b,c) for a,b,c in sets if 't' in a[:1]+b[:1]+c[:1]]
print(len(tset))
print(tset)
            
```

```python
def getConnections(iters=100):
    groups = set((c,) for c in computerset)
    computers = list(computerset)
    for _ in range(iters):
        print(_)
        newgroups = set()
        for candidate_computer in computers:
            for group in groups:
                found = True
                for group_computer in group:
                    if candidate_computer not in adj[group_computer]:
                        found = False
                        break
                if found:
                    members = [c for c in group] + [candidate_computer]
                    members.sort()
                    newgroups.add(tuple(members))
        if not newgroups:
            return groups
        groups = newgroups
res = getConnections()
print(res)              
                    
```
