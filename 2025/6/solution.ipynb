```python
import numpy as np
np.set_printoptions(suppress=True)
numbers = np.copy(numbers_)
symbols = np.copy(symbols_)
def tenmul(height, width):
    arr = np.zeros((0, width))   # start 2D
    for _ in range(height):
        arr = arr * 10
        new_row = np.ones((1, width))
        arr = np.vstack([arr, new_row])
    return arr
def exp_adjust(arr):
    h,w = arr.shape
    exps = np.log10(arr+1)
    exps = np.ceil(exps)
    maxexps = np.max(exps,axis=0)
    expdelta =  maxexps - exps
    adjexps = np.power(10,expdelta)
    adjarr = adjexps*arr
    return adjarr


h, w = numbers.shape
mul = tenmul(h, w)

mulnums = np.ones(w)      # shape (w,)
sumnums = np.zeros(w)     # shape (w,)

numbers = exp_adjust(numbers)
print(numbers)

while np.sum(numbers) != 0:
    print()
    digits = np.mod(numbers,10)                 # extract digit
    print(digits)
    col = np.sum(digits*mul, axis=0)
    print(col)
    sumnums += col
    mulnums *= col

    numbers = numbers // 10               # shift to next digit
    numbers = np.floor(numbers)
print()
print(sumnums)
print(mulnums)
print(symbols)
total = mulnums * ismuls + sumnums * isadds


totalsum = np.sum(total)
print(totalsum)
```
