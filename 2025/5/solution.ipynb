```python
import utils
from copy import deepcopy
inp = utils.get_input(2025,5,"input")
inp = [r.split('-') for r in inp]
ranges = [(int(r[0]),int(r[1])) for r in inp if len(r) == 2]
nums = [int(r[0]) for r in inp if len(r) == 1]
ranges.sort(key= lambda x: (x[0],-x[1]),reverse=True)
nums.sort(reverse=True)
```

```python
p1,p2 = 0,0
while ranges:
    l,r = ranges.pop()
    while ranges and ranges[-1][0] <= r:
        r = max(r,ranges.pop()[1])
    p2 += r - l + 1
    while nums and nums[-1] <= r:
        num = nums.pop()
        p1 += l <= num

print('p1',p1)
print('p2',p2)
```
