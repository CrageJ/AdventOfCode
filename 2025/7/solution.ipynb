```python

import utils
from copy import deepcopy
import numpy as np
inp = utils.get_input(2025,7,"input")
```

```python
from functools import cache

n,m = len(inp[0]),len(inp)

def getCell(x,y):
    if 0 <= x < n and 0 <= y < m:
        return inp[y][x]
    return "?"

def findS():
    for x in range(n):
        for y in range(m):
            if getCell(x,y) == "S":
                return x,y
    raise Exception("No S")

sx,sy = findS()
splits = set()

@cache
def iterate(x,y):
    cell = getCell(x,y)
    if cell == "?":
        return 1
    if cell == "^":
        splits.add((x,y))
        return iterate(x+1,y) + iterate(x-1,y)
    else:
        return iterate(x,y+1)

p2=iterate(sx,sy)
p1=len(splits)
print("p1", p1)
print("p2",p2)
```
