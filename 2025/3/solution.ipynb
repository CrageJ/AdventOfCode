```python
from bisect import bisect_left
from itertools import accumulate
from random import randint

from matplotlib import pyplot as plt

import utils
inp = utils.get_input(2025,3,"input")
inp = [r for r in inp]
```

```python
print(len(inp))
```

```python
from functools import cache
p1,p2 = 0,0
for s in inp:
    n = len(s)
    s = s[::-1]
    @cache
    def find(idx,depth):
        if idx >= n or depth == 0:
            return 0
        c = int(s[idx])
        take = c+10*find(idx+1,depth-1)
        notake = find(idx+1,depth)
        return max(take, notake)
    p1 += find(0,2)
    p2 += find(0,12)
print("p1",p1)
print("p2",p2)
```
