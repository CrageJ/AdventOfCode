```python
import utils
inp = utils.get_input(2025,1,"input")
inp = [("R"==row[0],int(row[1:])) for row in inp]
```

```python
loc = 50
n = 100
p1,p2= 0,0
for (rotation,magnitude) in inp:
    revolutions = magnitude//n
    magnitude %= n
    p2 += revolutions
    newloc = None
    if rotation:
        newloc = (loc + magnitude) % n
        if (newloc < loc and loc) or (loc!=0 and newloc ==0):
            p2 += 1
    else:
        newloc = (loc + n - magnitude) % n
        if (newloc > loc and loc) or (loc!=0 and newloc ==0):
            p2 += 1
    loc = newloc
    p1 += loc==0
print("p1",p1)
print("p2",p2)
```
